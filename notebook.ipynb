```python
import pandas as pd
import numpy as np
```

```python
# Loading dataset for chart 1: Line chart for country
co2= pd.read_csv("datasets/CO2.csv")
ch4= pd.read_csv("datasets/CH4.csv")
ghg= pd.read_csv("datasets/GHG.csv")
```

```python
# Filter by continent
continent = pd.read_csv('datasets/Countries by continents.csv')
```

```python
continent
```

```python
# Chart 2: Map for CO2 emmision
co2_emis= pd.read_csv("datasets/annual-co2-emissions-per-country/annual-co2-emissions-per-country.csv")
credit = pd.read_csv("datasets/share-co2-embedded-in-trade/share-co2-embedded-in-trade.csv")
co2_share= pd.read_csv("datasets/annual-share-of-co2-emissions/annual-share-of-co2-emissions.csv")
```

```python
co2_emis['Entity'].value_counts().sort_values()
```

```python
#chart 3: Stacked chart for electricity generation by source
electric = pd.read_csv("datasets/per-capita-electricity-source-stacked/per-capita-electricity-source-stacked.csv")
# chart 4: Area chart for energy comsumption
energy = pd.read_csv("datasets/energy consumption by source _ area chart/energy-consumption-by-source-and-country.csv")
# chart 5: full-stack for fuel, nuclear
sector = pd.read_csv("datasets/per-capita-electricity-fossil-nuclear-renewables.csv")
```

```python
sector[sector['Entity'] == 'Africa']
```

```python
# Comebine dataset
data = (
    co2
    .merge(ch4, on=['Entity', 'Year'], how='inner')
    .merge(ghg, on=['Entity', 'Year'], how='inner')
    .merge(co2_emis, on=['Entity', 'Year'], how='inner')
    .merge(credit, on=['Entity', 'Year'], how='inner')
    .merge(co2_share, on=['Entity', 'Year'], how='inner')
)
```

```python
data['Entity'].value_counts()
data.drop(['Code_x', 'Code_y', 'Code'], axis= 1, inplace= True)
```

```python
# Comebine dataset
full_data = (
    data
    .merge(electric, on=['Entity', 'Year'], how='left')
    .merge(energy, on=['Entity', 'Year'], how='left')
    .merge(sector, on=['Entity', 'Year'], how='left')
)
```

```python
full_data.drop(['Code_x', 'Code_y', 'Code'], axis= 1, inplace= True)
```
